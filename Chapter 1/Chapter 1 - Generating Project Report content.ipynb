## Functions to query ChatGPT, save chat history, and save responses to txt files

```python
def get_completion(prompt, message_history):
    messages = message_history.copy()
    messages.append({"role": "user", "content": prompt})

    # Temperature left as default to get similar results in the api as the chat interface.
    # Prompts have been tested in the chat interface.
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages
    )
    response_content = response['choices'][0]['message']['content']
    messages.append({"role": "assistant", "content": response_content})
    print('Prompt: ', prompt)
    print('Response: ', response_content)
    return response_content, messages
```

```python
def save_assistant_messages(message_history, file_name):
    # Specify the folder where the text files will be saved
    folder = 'H:\\Documents\\Software Development\\QUB Software Development\\Data_analysis_module\\AITxtResponses\\Ch1AssistantResponseTxtFiles'
    
    # Combine the folder path and file name to get the full file path
    file_path = f'{folder}\\{file_name}.txt'
    
    # Open the text file in write mode
    with open(file_path, 'w') as file:
        # Iterate through each message in the history
        for message in message_history:
            role = message['role']
            content = message['content']
            
            # Only write the assistant messages to the file
            if role == 'assistant':
                # Write each message to the file in the format: Assistant: content
                file.write(f'Assistant: {content}\n\n')

# Example usage:
# save_assistant_messages(message_history, 'my_document_name')
```

## 1.3.1. Value

```python
# Clearing message history to avoid hitting the token limit
message_history = []
```

*Firstly, a list of values is generated and saved in JSON format whcih will later be looped over and queried for deeper information.*

*The number_of_values can be changed depending on how many you want to investigate, but here I've restricted it to 2 in the interest of keeping the report a reasonable length.*

```python
# Set the number of values you want the AI to generate
number_of_values = 2;
prompt = f"""Imagine you are an expert business analyst who has spent decades studying companies 
            in the {industry} industry. 
            In your experience, what have successful businesses in the industry focused on and optimised 
            in order to become successful and maximise revenue?  
            Provide a prioritised, numbered list in json format of the {number_of_values} most important factors. 
            Factors to consider could include company values, KPIs, innovation, customer focus, 
            talent management, or other theories of business success you’ve studied. 
            The format should be {{
                "factors": [
                    {{
                        "priority": 1,
                        "factor": " ",
                        "description": " "
                    }},
                    ...
                ]
            }}"""
```

```python
completion, message_history = get_completion(prompt, message_history)
```

```python
save_assistant_messages(message_history, 'Values')
```

```python
values_dict = json.loads(completion)
```

```python
print(values_dict)
```

```python
# Clearing message history to avoid hitting the token limit
message_history = []
```

*Here's where we're looping through the list of values and asking a series of questions about each value. Mainly I'm interested in the metrics and causal factors for each value.*

*Information on each value is saved to a unique txt file - Value1, Value2 etc.*

```python
for i, factor in enumerate(values_dict['factors']):
    factor_name = factor['factor']
    factor_description = factor['description']

    # generate prompt for Q1
    prompt_q1 = f"""Imagine you are an expert business analyst who has spent decades studying companies 
                in the {industry} industry.
                As a business analyst you have had to measure how specific factors have contributed to the 
                success or failure of businesses in the industry over previous decades.
                A metric refers to a quantitative measure used to assess or evaluate a particular phenomenon, 
                system, or process.
                Metrics are objective, measurable, and can be expressed numerically.
                Provide a list of metrics that you measured that led you to the conclusion that the 
                following factor was important for the success of businesses in the industry:
                Factor: {factor_name}
                Description: {factor_description}"""

    # call your API function for Q1
    completion, message_history = get_completion(prompt_q1, message_history)

    # generate prompt for Q2
    prompt_q2 = f"""A causal factor refers to an element or variable that contributes to the cause or 
                explanation of an event, outcome, or phenomenon. They cause a metric to go up or down.
                Here’s an example showing the difference between a metric and a causal factor:
                The average test score is a metric that measures the performance of a group of students on a 
                particular exam.
                Study time can be considered a causal factor that influences the average test score. The 
                amount of time students dedicate to studying directly affects their level of preparedness 
                and knowledge retention, which, in turn, impacts their test scores. Generally, as students 
                increase their study time, their average test scores are likely to improve.
                For each of the metrics you outlined, tell me the causal factors that cause the metric to be 
                higher or lower."""

    # call your API function for Q2
    completion, message_history = get_completion(prompt_q2, message_history)
    
    # Generate prompt for Q3
    # GPT3.5 struggles to generate real links, but testing in the 4.0 chat worked better online.
    prompt_q3 = f"""As an expert business analyst you have gained
                knowledge from many different data sources including websites.
                I would like to explore each metric further. 
                Could you provide a URL or source 
                where I can find more information 
                about each metric you mentioned?
                I need real data for each metric."""
    
    # call your API function for Q2
    completion, message_history = get_completion(prompt_q3, message_history)
    
    
    # save the metrics and causes for each value
    save_assistant_messages(message_history, f'Value{i+1}')
    
    # clear message history so that an unlimited number of values can be looped without hitting token limit
    message_history = []
```
