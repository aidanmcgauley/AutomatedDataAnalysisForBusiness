**"I now want to clean the 'Current Year' and Equity(CUR) columns. How do i remove any rows that either don't contain data for the year or for equity?"**

You can remove rows that contain missing data for either 'Current Year' or 'Equity(CUR)' by using the dropna() function in pandas. By specifying a subset of columns to consider, dropna() will remove any row with missing data in those columns.

```python
# Start with your original DataFrame
data_clean2 = data.dropna(subset=['Current Year', 'Equity(CUR)'])  # drop rows where 'Current Year' or 'Equity(CUR)' are NaN

data_clean3 = data_clean2[data_clean2['Equity(CUR)'] != 1.0]  # from the result, drop rows where 'Equity(CUR)' equals 1.0
```

**"Now that I've cleaned the Current Year and Equity columns, I want to display them in a graph. What's the best way to do this considering there will be several rows for each company since equity is recorded by year?"**

For visualizing the 'Current Year' and 'Equity(CUR)' columns considering that there are several rows for each company, a line plot would be a great option. With each company being a separate line, you can observe the change in equity over time for each company.

Here's an example using matplotlib:

Note: If you have a large number of companies, the legend and the plot can become quite cluttered. It may be better to plot a subset of companies, or to not include a legend, if this is the case.

This approach will create one line per company. Each line will represent the change in equity for that company over the years.

Keep in mind that if you have many companies this plot can become cluttered and hard to read. A more scalable solution could involve interactive plotting libraries like Plotly, which would allow you to hover over lines to see which company they represent.

```python
# create a figure and axis
fig, ax = plt.subplots()

# for each unique company, plot a line
for company in data_clean3['Company Name'].unique():
    company_data = data_clean3[data_clean3['Company Name'] == company]
    ax.plot(company_data['Current Year'], company_data['Equity(CUR)'], label=company)

# set the labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Equity(CUR)')
ax.set_title('Equity over time by company')

# optionally add a legend, but be careful if you have a lot of companies!
# ax.legend()

# show the plot
plt.show()
```

If the data for 2020 is NaN, matplotlib will skip this point and connect the line between 2019 and 2021 directly. You can use the same approach as above to find the number of employees for the most recent year of each company. Then, you can use this data to create a bar plot. Here's how you can do it:

This code will create a bar plot where the x-axis represents the company names and the y-axis represents the number of employees in the most recent year of available data for each company.
```python
import seaborn as sns

# Identify the row with the most recent year for each company
recent_rows = data.loc[data.groupby('Company Name')['Current Year'].idxmax()]

# Create a bar plot
plt.figure(figsize=(10,6))
sns.barplot(x='Company Name', y='Employees(CUR)', data=recent_rows)

# Rotate x-labels for better visibility
plt.xticks(rotation=90)

# Set title and labels
plt.title('Number of Employees in Most Recent Year by Company')
plt.xlabel('Company Name')
plt.ylabel('Number of Employees')

# Display the plot
plt.show()
```

**"I don't want to include companies in the chart if there is no available data for the number of employees"**

You can add a step to filter out companies with no available data for the number of employees before creating the plot:

This code will first remove any companies that have NaN values in the 'Employees' column for the most recent year of available data, and then plot the remaining data.

```python
# Identify the row with the most recent year for each company
recent_rows = data.loc[data.groupby('Company Name')['Current Year'].idxmax()]

# Filter out companies with no data for number of employees
recent_rows = recent_rows[recent_rows['Employees(CUR)'].notna()]

# Create a bar plot
plt.figure(figsize=(10,6))
sns.barplot(x='Company Name', y='Employees(CUR)', data=recent_rows)

# Rotate x-labels for better visibility
plt.xticks(rotation=90)

# Set title and labels
plt.title('Number of Employees in Most Recent Year by Company')
plt.xlabel('Company Name')
plt.ylabel('Number of Employees')

# Display the plot
plt.show()
```

**"I also want to clean up the 'No. Officers' column.  It's currently in the format '5 officers  / 3 resignations', but I just want to keep the first number as an int. You'll need to account for blank values."**

For the 'No. Officers' column, we can use the str.split() function to split the string at the first space, which will separate the number of officers from the rest of the string. Then we can use str[0] to select the first element of the resulting split string (which is the number of officers). Here's how you can do this:

For the 'No. Officers' column, you can handle cases where the first part of the string is not a number in a similar way to employees. If the first part of the string is always supposed to be a number, you can remove rows where this is not the case:

```python
print(data['No. Officers'])
```

```python
# Extract the first part of the string in 'No. Officers'
data['No. Officers'] = data['No. Officers'].str.split(' ').str[0]

# Remove rows where this is not a number
data = data[pd.to_numeric(data['No. Officers'], errors='coerce').notnull()]

# Convert 'No. Officers' to int
data['No. Officers'] = data['No. Officers'].astype(int)
print(data['No. Officers'])
```
